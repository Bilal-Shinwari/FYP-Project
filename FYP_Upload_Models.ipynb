# FYP Model Uploader — Run this ONCE
This notebook uploads your local model files to Google Drive so `FYP_Colab_Backend.ipynb` can use them.

**You need to upload these files from your PC:**
| File | Location on your PC | Size |
|------|--------------------|----- |
| `t3_finetuned.safetensors` | `D:/FYP/FYP/t3_finetuned_model/chatterbox-finetuning/` | 2 GB |
| `t3_cfg.safetensors` | `D:/FYP/FYP/t3_finetuned_model/chatterbox-finetuning/pretrained_models/` | 2 GB |
| `s3gen.safetensors` | same pretrained_models folder | 1 GB |
| `ve.safetensors` | same pretrained_models folder | 5.5 MB |
| `conds.pt` | same pretrained_models folder | 105 KB |
| `tokenizer.json` | same pretrained_models folder | 25 KB |
| `2.wav` | `.../speaker_reference/` | 936 KB |

**Run each cell one by one. A file picker will open for each upload.**

```
# ── Step 1: Mount Google Drive ────────────────────────────────────────────
from google.colab import drive
drive.mount('/content/drive')

import os

# Create all folders
folders = [
    '/content/drive/MyDrive/FYP_Models',
    '/content/drive/MyDrive/FYP_Models/pretrained_models',
    '/content/drive/MyDrive/FYP_Models/speaker_reference',
]
for f in folders:
    os.makedirs(f, exist_ok=True)

print('✅ Google Drive mounted.')
print('✅ Folder structure created:')
print('   MyDrive/FYP_Models/')
print('   MyDrive/FYP_Models/pretrained_models/')
print('   MyDrive/FYP_Models/speaker_reference/')
```

```
# ── Step 2: Upload t3_finetuned.safetensors (2 GB) ────────────────────────
# A file picker will open — navigate to:
# D:/FYP/FYP/t3_finetuned_model/chatterbox-finetuning/t3_finetuned.safetensors

from google.colab import files
import shutil

DEST = '/content/drive/MyDrive/FYP_Models/t3_finetuned.safetensors'

if os.path.exists(DEST):
    print('✅ Already on Drive — skipping upload.')
else:
    print('📂 File picker opening... select t3_finetuned.safetensors')
    uploaded = files.upload()
    fname = list(uploaded.keys())[0]
    shutil.move(fname, DEST)
    size = os.path.getsize(DEST) / 1e9
    print(f'✅ Saved to Drive ({size:.2f} GB)')
```

```
# ── Step 3: Upload t3_cfg.safetensors (2 GB) ──────────────────────────────
# Select: D:/FYP/FYP/t3_finetuned_model/chatterbox-finetuning/pretrained_models/t3_cfg.safetensors

DEST = '/content/drive/MyDrive/FYP_Models/pretrained_models/t3_cfg.safetensors'

if os.path.exists(DEST):
    print('✅ Already on Drive — skipping.')
else:
    print('📂 File picker opening... select t3_cfg.safetensors')
    uploaded = files.upload()
    fname = list(uploaded.keys())[0]
    shutil.move(fname, DEST)
    size = os.path.getsize(DEST) / 1e9
    print(f'✅ Saved to Drive ({size:.2f} GB)')
```

```
# ── Step 4: Upload s3gen.safetensors (1 GB) ───────────────────────────────
# Select: .../pretrained_models/s3gen.safetensors

DEST = '/content/drive/MyDrive/FYP_Models/pretrained_models/s3gen.safetensors'

if os.path.exists(DEST):
    print('✅ Already on Drive — skipping.')
else:
    print('📂 File picker opening... select s3gen.safetensors')
    uploaded = files.upload()
    fname = list(uploaded.keys())[0]
    shutil.move(fname, DEST)
    size = os.path.getsize(DEST) / 1e6
    print(f'✅ Saved to Drive ({size:.0f} MB)')
```

```
# ── Step 5: Upload small files (ve.safetensors, conds.pt, tokenizer.json, 2.wav)
# Select ALL 4 files at once in the file picker (hold Ctrl to multi-select)

DEST_MAP = {
    've.safetensors'  : '/content/drive/MyDrive/FYP_Models/pretrained_models/ve.safetensors',
    'conds.pt'        : '/content/drive/MyDrive/FYP_Models/pretrained_models/conds.pt',
    'tokenizer.json'  : '/content/drive/MyDrive/FYP_Models/pretrained_models/tokenizer.json',
    '2.wav'           : '/content/drive/MyDrive/FYP_Models/speaker_reference/2.wav',
}

missing = [k for k, v in DEST_MAP.items() if not os.path.exists(v)]

if not missing:
    print('✅ All small files already on Drive — skipping.')
else:
    print(f'📂 File picker opening... select these {len(missing)} files (hold Ctrl for multi-select):')
    for f in missing:
        print(f'   - {f}')
    uploaded = files.upload()
    for fname, data in uploaded.items():
        if fname in DEST_MAP:
            with open(DEST_MAP[fname], 'wb') as out:
                out.write(data)
            print(f'  ✅ {fname} saved')
        else:
            print(f'  ⚠ Unexpected file: {fname} — ignored')
```

```
# ── Step 6: Verify everything is in place ────────────────────────────────
expected = [
    '/content/drive/MyDrive/FYP_Models/t3_finetuned.safetensors',
    '/content/drive/MyDrive/FYP_Models/pretrained_models/t3_cfg.safetensors',
    '/content/drive/MyDrive/FYP_Models/pretrained_models/s3gen.safetensors',
    '/content/drive/MyDrive/FYP_Models/pretrained_models/ve.safetensors',
    '/content/drive/MyDrive/FYP_Models/pretrained_models/conds.pt',
    '/content/drive/MyDrive/FYP_Models/pretrained_models/tokenizer.json',
    '/content/drive/MyDrive/FYP_Models/speaker_reference/2.wav',
]

all_ok = True
print('Final check:')
for path in expected:
    exists = os.path.exists(path)
    size = f'{os.path.getsize(path)/1e6:.1f} MB' if exists else 'MISSING'
    icon = '✅' if exists else '❌'
    print(f'  {icon}  {os.path.basename(path):30s} {size}')
    if not exists:
        all_ok = False

print()
if all_ok:
    print('🎉 ALL FILES READY!')
    print('   You can now open FYP_Colab_Backend.ipynb and run the backend.')
else:
    print('⚠ Some files are missing — re-run the cells above for those files.')
```
